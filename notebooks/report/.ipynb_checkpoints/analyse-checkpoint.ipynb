```python
import pandas
```

```python
df = pandas.read_excel('data/flowers.xlsx')
df.head()
```

```python
line_filter = df['Lijn'] == 'Spray Carnations'
column_filter = ['Date', 'Sales (qty)']

df[line_filter][column_filter].set_index('Date').plot()
```

```python
date_filter = (df['Date'] > '2016') & (df['Date'] < '2017')

df[line_filter & date_filter][column_filter].set_index('Date').plot()
```

```python
date_filter = (df['Date'] > '2016-02') & (df['Date'] < '2016-04')

df[line_filter & date_filter][column_filter].set_index('Date').plot()
```

https://www.timeanddate.com/holidays/uk/2016

6 maart 2016 was een moederdag in UK

```python
df['dayofweek'] = df['Date'].dt.dayofweek
column_filter = ['Date', 'Sales (qty)', 'dayofweek']

df[line_filter & date_filter][column_filter].groupby('dayofweek')['Sales (qty)'].mean().plot()
```

The day of the week with Monday=0, Sunday=6.
